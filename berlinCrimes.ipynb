# Union Columnas


```python
# Union columnas para no tener tantas categorias
df["Robo violencia"] = df["Robbery"] + df["Street_robbery"]
df["Daño material"] = df["Damage"] + df["Graffiti"]
df["Incendio"] = df["Fire"] + df["Arson"]
df["Robo sin violencia"] = df["Theft"] + df["From_car"]
df["Robo vehiculos"] = df["Car"] + df["Bike"]
# Comprobamos que columnas den resultados correctos, que si estan ok
```

```python
# Eliminamos columnas innecesarias
df.drop(columns=["Robbery", "Street_robbery", "Damage", "Graffiti", "Fire", "Arson", "Theft", "From_car", "Car", "Bike", "Code"], inplace=True)
```

```python
df.rename(columns= {"Year": "Año", "District": "Distrito", "Location": "Ubicacion", "Injury": "Agresion leve", "Agg_assault": "Agresion grave", "Threat": "Amenaza/Rapto", "Burglary": "Robo con allanamiento", "Drugs": "Drogas"}, inplace=True)
```

```python
# Cambiamos orden de columnas
nuevo_orden = ["Año", "Distrito", "Ubicacion", "Robo sin violencia", "Robo violencia", "Robo con allanamiento", "Robo vehiculos", "Agresion leve", "Agresion grave", "Amenaza/Rapto", "Drogas", "Daño material", "Incendio", "Local"]
berlin = df[nuevo_orden]
```
